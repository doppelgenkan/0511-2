# **作業のルール**

この作業のルールを決めておきましょう。このルールは、君たちが作業を進めやすく、後でノートブックを見直したときに何をやったかを確認しやすくするためと、僕が提出された君たちの**ノートブック**をチェックするのに便利にするためです。  
- Rで作業を行うWebページ(例えばこのWebページ)を**ノートブック(Notebook)**といいます。
- クリックすると影付きの長方形で囲まれる領域を**セル**といいます。
- □で始まるセル(行)の文章は、その下の(入力)セルに何かを入力することにします。  
- 一番左に**`[  ]    1`**表示されていて、**灰色の長方形の枠**のあるセルが**入力セル**で、ここに数値、文字列、数式等を入力します。
 - 入力後は、**【Shift】＋【Enter】**を押して、処理を実行します。
 - 処理の実行結果は、入力セルのすぐ下に出力されます。
 - １つのセルには複数の行を入力することができます。その場合、１つの行の入力が済んだら**【Enter】キーを押すと次の行の入力**ができます。各行の左端には行番号がつきます（下リスト参照）。
```
1 df <- read.csv("/content/table-data.csv")
2 print(df)
3 x <- df$ubound
4 barplot(x, names=c("low", "mid", "high"))
```
> ※）上のリストは１つのセル内の４行のコマンドです。一番左にある数字，1, 2, 3, 4が行番号です。
- セルの下に新しいセルを挿入するには、その**セルをクリックしてツールバーの［＋］ボタンをクリック**します。［＋］ボタンは次の２種類あります：
  - ［+ コード］ はRのコマンドを入力するためのセルが挿入されます。
  - ［+ テキスト］ は文章などのテキストを入力するためのセルが挿入されます。
- > で始まるセルは、君たちの入力結果の**コメント**を示すことにします。
- 上記以外のセルは文章のセルで、入力や実行を行うセルではありません。その文章を読んでください。
- ノートブックは自動保存されますが、念のため、作業が終わったらNotebookの［ファイル］メニューから［保存］を選択して下さい(下図)。

## **レシピ4.8　CSVファイルから読み込む**(p.97)

統計解析では、データが必要です。データはノートブック上で入力することもできますが、非常に面倒で、このことがRの１つの欠点だと思われています。しかし、これは間違いで、データの入力はMicrosoft Excel, Apple Numbers, Google Spreadseet, LibreOffice Calcなど，優れた表計算ソフト等で行い，そのデータファイルを読み込めばよいのです（しかもExcel以外は無料です）。また，種々の機器を用いた計測では，その機器が出力する計測データを読み込まなければなりません。このセクションでは，多くのデータファイルに利用されているCSVファイル（[CSVファイルについてはこのリンクを見て下さい](https://www.pc-master.jp/sousa/csv.html)）の読み込みを練習してみます。ただ，オンラインのColaboratryでは教科書に書かれていない一手間が必要です。そのことを含めて解説していきましょう。  

まず，Cloud上に必要な**CSVファイルをアップロード**する必要があります。以下の手順で行って下さい。なお，データは自分のコンピュータ上の「data」フォルダにあるとします（あらかじめダウンロードしておいて下さい）。

1. 自分のコンピュータにある「data」フォルダを開いておいて下さい。
2. ノートブックの左端の縁にあるスクリーンショット 2020-09-21 15.18.49.pngボタンをクリックして下さい。そうするとページ左に**「ファイル」ペイン**が開きます。  
> 「ファイル」ペインには「sample_data」フォルダのみが表示されているはずです。
3. 開いている「data」フォルダから必要なファイルをノートブックの**「ファイル」ペインの下部**にドラッグして下さい。  
> このとき，「ファイルセッション　ストレージにアップロードするには、ここにドロップして下さい」が表示されますので，その部分に必要なファイルをドラッグ&ドロップします。  
4. 「注: アップロードしたファイルはランタイムのリサイクル時に削除されます。」というダイアログボックスが表示される場合は、［OK］ボタンをクリックします。  
5. 「ファイル」ペインに必要なファイルが表示されているはずです。これで，データファイルをノートブックで利用することができます。  
> **重要な注意**：ノートブックを一度終了すると、アップロードされたデータファイルは消去されます。したがって、あらためてノートブックで作業する場合は、上記1〜5を繰り返して下さい。
